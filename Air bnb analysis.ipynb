# Data Cleaning

```python
df.isnull().sum()
```

```python
#dropping missing values
df.dropna(inplace = True)
```

```python
#dealing with duplicates
df.duplicated().sum()

df[df.duplicated()]
```

```python
#deleted all duplicated records

df.drop_duplicates(inplace= True)
```

```python
df.duplicated().sum()
```

```python
# changing datatypes

df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
```


# Data Analysis 
**Univariate Analysis**

```python
#price distributions
df['price']
```

```python
#identifying outliers in price
dff = df[df['price'] < 1500]
sns.boxplot(data= dff, x= 'price')
```

```python
plt.figure(figsize = (8,5))
sns.histplot(data = dff, x = 'price' , bins = 100)
plt.ylabel("Frequency")
plt.title("Price Distribution")
plt.show()
```

```python
dff.columns
```

```python
df.dtypes
```

```python
plt.figure(figsize = (6,3))
sns.histplot(data = dff, x = 'availability_365')
plt.ylabel("Frequency")
plt.title("availability_365 Distribution")
plt.show()
```

```python
df.dtypes
```

```python
#average price per bed
dff.groupby(['neighbourhood_group'])['price per bed'].mean()
```

# feature engineering

```python
dff['price per bed']= dff['price']/dff['beds']
```

```python
dff.head()
```

# Bivariate Analysis

```python
dff.columns
```

```python
#price dependencies on neighbourhood
sns.barplot(data= dff, x= 'neighbourhood_group' , y= 'price', hue = 'room_type' )
```

```python
# number of reviews by price
plt.figure(figsize  = (8,5))
plt.title("Locality by Review")
sns.scatterplot(data = dff, x= 'number_of_reviews', y= 'price', hue= 'neighbourhood_group')
plt.show()
```

```python
dff.dtypes
```

```python
sns.pairplot(data= dff, vars= ['price', 'minimum_nights', 'number_of_reviews', 'availability_365'], hue = 'room_type')
```

```python
#Geographical distributions AirBnB

sns.scatterplot(data=dff, x= 'longitude', y= 'latitude', hue = 'room_type')
plt.title("Geographical Distribution of AirBnB")
plt.show()
```

```python
dff.dtypes
```

```python
# heat map - correlation of one variale with others for numerical column

corr = dff[['latitude', 'longitude', 'price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'availability_365', 'beds']].corr()
corr
```

```python
plt.figure(figsize = (8,6))
sns.heatmap(data= corr, annot=True)
```
